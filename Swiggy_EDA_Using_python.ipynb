### 6.1 Rating Distribution

```python
plt.figure(figsize=(8,4))
plt.hist(df["Rating"],bins=20,color="lightgreen",edgecolor="black")
plt.title("Rating Distribution")
plt.xlabel("Rating Count")
plt.ylabel("Frequency")
plt.show()
```

insights:

### 6.2 Price Distribution

```python
#2.Price Distribution
plt.figure(figsize=(6,4))
sns.histplot(df["Price (INR)"],bins=30,color="green",edgecolor="black",kde=True)
plt.title("Price Distribution")
plt.xlabel("Price (INR)")
plt.show()
```

## Insights

```python
Insights: Price Distribution
#1.The majority of restaurants fall within the â‚¹200â€“â‚¹600 price range,
#2.suggesting that Swiggyâ€™s offerings target affordable and mid-tier customers.
#3.High-priced premium restaurants are comparatively rare.
```

### 6.3 price Range

```python
plt.figure(figsize=(6,5))
sns.violinplot(data=df,x="Price (INR)")
plt.title("price Range", fontsize=14)
plt.xlabel("State", fontsize=12)
plt.ylabel("Price Range", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

## Insights:

```python
insihts:Price Range
1. Most prices are near 0: High density of prices concentrated near 0.
2. Long tail: Prices can extend to low values (around -7000).
3. Asymmetric distribution: Bulk of data near 0, spreading out unevenly.
4. Variation exists: Prices vary widely acrossÂ theÂ dataset.
```

# 6.2 Bivariate Analysis

## 6.2.1 Average Rating by Price Range

```python
avg_rating = df.groupby('Price (INR)')['Rating'].mean().reset_index()
avg_rating.rename(columns={'Price (INR)': 'Price Range'}, inplace=True)

plt.figure(figsize=(10,5))
sns.lineplot(x='Price Range', y='Rating', data=avg_rating, marker='o', color='royalblue')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

## insights:

```python
insights:  Average Rating by Price Range
#1.As the price increases, the average rating also tends to rise slightly.
#2.This suggests that higher-priced restaurants are often rated better, indicating better quality or customer satisfaction.
#3.However, extremely high-priced options show limited rating improvement â€” implying diminishing returns on cost vs satisfaction.
```

# 6.3 multivariate Analysis

### 6.3.1 Correlation Between Numeric Features

```python
plt.figure(figsize=(5,3))
sns.heatmap(df[['Price (INR)', 'Rating', 'Rating Count']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Numeric Features')
plt.show()
```

 ## Insights

```python
insights:Correlation Between Numeric Features
#1.Rating Count and Rating show a positive correlation (0.26) â€” restaurants with more reviews often have slightly higher ratings.
#2.Price (INR) has a weak negative correlation with both Rating (-0.1) and Rating Count (-0.1) â€” higher prices donâ€™t guarantee better ratings or more reviews.
#3.Overall, there is no strong correlation among numeric features, indicating they are mostlyÂ independent.
```

### 6. 3.2 Pairwise Relationships Between Variables

```python
sns.pairplot(df[['Price (INR)', 'Rating', 'Rating Count']], diag_kind='kde')
plt.suptitle('Pairwise Relationships Between Variables', y=1.02)
plt.show()
```

## Insights

```python
insights:Pairwise Relationships Between Variables
#1.Rating Count vs Rating: Slight positive relationship â€” restaurants with more reviews tend to have higher ratings.
#2.Price vs Rating: Weak or no visible relationship â€” higher price doesnâ€™t ensure better ratings.
#3.Price vs Rating Count: Very weak or no correlation â€” expensive restaurants donâ€™t always get more reviews.
#4.Most data points are clustered in lower price ranges and high ratings (4â€“5), showing that affordable and well-rated restaurantsÂ dominate.
```
